```python
import pandas as pd
from tqdm import tqdm
import ast
tqdm.pandas()
```

```python
def unstack(df):
    df.breeding_results = df.breeding_results.apply(lambda x: ast.literal_eval(x))
    df.breeding_results_color = df.breeding_results_color.apply(lambda x: ast.literal_eval(x))
    df.probability = df.probability.apply(lambda x: ast.literal_eval(x))
    breeding_pair_list = df.breeding_pair.tolist()
    breeding_pair_color_list =  df.breeding_pair_color.tolist()
    breeding_results_list = df.breeding_results.tolist()
    breeding_results_color_list = df.breeding_results_color.tolist()
    probability_list = df.probability.tolist()
    arr=[]
    for ind, breed_results in enumerate(breeding_results_list):
        breeding_pair = breeding_pair_list[ind]
        breeding_pair_color = breeding_pair_color_list[ind]
        breed_res_colors = breeding_results_color_list[ind]
        breed_res_prob = probability_list[ind]
        for sub_ind, breed_res in enumerate(breed_results):
            arr.append([breeding_pair, breeding_pair_color, breed_res, breed_res_colors[sub_ind], breed_res_prob[sub_ind]])
            
    return pd.DataFrame(arr,columns = ['breeding_pair', 'breeding_pair_color', 'result', 'result_color','probability'])
```

```python
flowers = ['rose', 'cosmo','lily','pansy','hyacinth','tulip','mum','windflower']
```

```python
flower_genes_table = {}
flower_genes = {}
for flower in flowers:
    flower_genes_table[flower] = unstack(pd.read_csv('genes_table/'+flower+'_genes_table.csv'))
    flower_genes_df = pd.read_csv('genes_table/'+flower+'_all_genes.csv')
    flower_genes_df['note'] = flower_genes_df.color.apply(lambda x: 'seed' if 'seed' in x else '')
    flower_genes_df['color'] = flower_genes_df.color.apply(lambda x: x.split(' ')[0])
    flower_genes[flower] = flower_genes_df
```
